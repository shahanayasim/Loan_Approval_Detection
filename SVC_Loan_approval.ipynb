```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/datasets/LoanApprovalPrediction.csv")
df
```

```python
lst=['Gender','Married','Dependents','Self_Employed',"Loan_Amount_Term","Credit_History"]
for i in lst:
  df[i]=df[i].fillna(df[i].mode()[0])
```

```python
df.isna().sum()
```

```python
df['LoanAmount']=df["LoanAmount"].fillna(df["LoanAmount"].mean())
```

```python
df.isna().sum()
```

```python
#since loan_id is not needed ,we remove that column
#also convert object type into int type
#map()
df["Gender"]=df["Gender"].map({'Male':0,'Female':1})
df
```

```python
df["Married"]=df['Married'].map({'Yes':1,'No':0})
df
```

```python
df["Education"]=df['Education'].map({'Graduate':1,'Not Graduate':0})
df
```

```python
df["Self_Employed"]=df['Self_Employed'].map({'Yes':1,'No':0})
df
```

```python
df['Dependents'].unique()
```

```python
df['Dependents']=df['Dependents'].map({'0':0,'1':1,'2':2,'3+':3})
df
```

```python
df["Property_Area"]=df['Property_Area'].map({'Urban':1,'Rural':0,'Semiurban':2})
df
```

```python
df["Loan_Status"]=df['Loan_Status'].map({'Y':1,'N':0})
df
```

```python
df.drop(['Loan_ID'],axis=1,inplace=True)
df
```
